## Create the best plot using polynomials!

Below, we created a plot with a clearly non-linear shape.

- plot a polynomial function using `PolynomialFeatures` for polynomials up until the second, third and fourth degree.
- print out the $R^2$ value for each of the three results. Draw conclusions with respect to which degree is best.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

def pol(x):
    return x * np.cos(x)

x = np.linspace(0, 12, 100)
rng = np.random.RandomState(1234)
rng.shuffle(x)
x = np.sort(x[:25])
y = pol(x) + np.random.randn(25)*2

plt.scatter(x, y, color='green', s=50, marker='.')

plt.show();
```

## Solution

```python
## your code here
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
x = x[:, np.newaxis]
X_linspace = np.linspace(0,12,100)
X_linspace = X_linspace[:,np.newaxis]
plt.figure(figsize=(12,12))
plt.scatter(x,y,color='green')
for index,j in enumerate([2,3,4]):
    poly=PolynomialFeatures(j)
    X=poly.fit_transform(x)
    X_linspace_fin = poly.fit_transform(X_linspace)
    reg_poly=LinearRegression().fit(X,y)
    predict_x=reg_poly.predict(X)
    y_poly_pred = reg_poly.predict(X_linspace_fin)
    score=r2_score(y,predict_x)
    plt.plot(X_linspace,y_poly_pred, label='degree %d' %j+' r^2: ' +str(round(score,2)))
plt.legend()
plt.show()
```
